```python
nltk.download('stopwords')
stopwords = stopwords.words('english')
```

```python
swdr_neg = set(stopwords) - set(['no','nor','not','never','against'])
```

```python
swdr_neg
```

```python
def remove_stp(txt):
    st=""
    for val in list(txt.split(" ")):
        if val not in swdr_neg:
            st=st+val+" "
    return st
        
```

```python
df['splch_rem1'] = df['splch_rem'].apply(remove_stp)
```
