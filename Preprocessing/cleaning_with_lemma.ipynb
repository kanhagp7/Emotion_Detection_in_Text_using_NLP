```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
import emoji
import re
from langdetect import detect
import nltk
from nltk.stem import WordNetLemmatizer
import contractions
```

```python
df = pd.read_csv("label3.csv",encoding='utf-8',encoding_errors='ignore')
df.head(40)
```

```python
len(df)
```

```python
df.info()
```

```python
df['Comments']=df['Comments'].str.lower()
```

```python
def remove_emojis(text):
    return emoji.replace_emoji(text, replace="")
```

```python
df['without_emoji']= df['Comments'].apply(remove_emojis)
```

```python
df['without_emoji'].info()
```

```python
df.head(40)
```

```python
#Converting Contractions
```

```python
df['cont_rem'] = df['without_emoji'].apply(contractions.fix)
```

```python
df.head(60)
```

```python
#Remving Numbers
```

```python
def remove_num(comm):
    return re.sub(r"[0-9]", "", comm).strip()
```

```python
df['num_rem']= df['cont_rem'].apply(remove_num)
```

```python
df.tail(60)
```

```python
# Removing Hashtags and @
```

```python
def remove_tags(comm):
    return re.sub(r'[@#]\S+', '', comm).strip()
```

```python
df['without_tags'] = df['num_rem'].apply(remove_tags)
```

```python
df.head(60)
```

```python
# Remove special charachters and puntuatobs
```

```python
def remove_splch(comm):
    return re.sub(r'[^A-Za-z0-9\s]', ' ', comm)
```

```python
df['splch_rem'] = df['without_tags'].apply(remove_splch)
```

```python
df.head(60)
```

```python
dfnew = df[df['splch_rem'].apply(lambda x:  5<= len(x.split()) <= 70)]
```

```python
from transformers import pipeline
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base",return_all_scores=True, 
                      truncation=True, 
                      max_length=512)
classifier("I love this!")
```

```python
sk=classifier("In Germany they spit on their Veterans A disgrace")
```

```python
sk[0]
```

```python
def getlabel(text):
    sk =classifier(text)
    max=0
    lb = 'neutral'
    for val in sk[0]:
        if val['score']>max:
            max = val['score']
            lb=val['label']
    return lb
```

```python
ck = getlabel(" trump is such a stupid ")
print(ck)
```
